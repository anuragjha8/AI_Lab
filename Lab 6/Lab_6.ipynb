```python
def grad(a,b,c):
  w=np.linspace(-12,12,num=100)
  L=[]
  Ls=[]
  for i in range(len(w)):
    L.append((a/2)*(w[i]**2) + b*w[i] + c)
  plt.plot(w,L,color='red')
  plt.scatter((-b/a),c-((b*b)/(2*a)),color='blue')
  plt.scatter((-b/a),0,color='green',marker='*')
```

```python
def grades(a,b,c,d,e,lr,t):
  w=np.linspace(-4,4,num=100)
  w0=-3
  L=[]
  for i in range(len(w)):
    L.append(a*(pow(w[i],4)) + b*(pow(w[i],3)) + c*(w[i]**2) + d*w[i] + e)
  plt.plot(w,L,color='red')
  for i in range(t):
    dL=4*a*w0*w0*w0 + 3*b*w0*w0 + 2*c*w0 + d
    plt.scatter(w0,a*w0*w0*w0*w0 + b*w0*w0*w0 + c*w0*w0 + d*w0 + e,color='black')
    w0=w0 - lr*dL
  plt.scatter(w0,a*w0*w0*w0*w0 + b*w0*w0*w0 + c*w0*w0 + d*w0 + e,color='yellow')
```

```python
def moment(a,b,c,d,e,lr,beta,t):
  w=np.linspace(-4,4,num=100)
  w0=-3
  L=[]
  for i in range(len(w)):
    L.append(a*(pow(w[i],4)) + b*(pow(w[i],3)) + c*(w[i]**2) + d*w[i] + e)
  plt.plot(w,L,color='red')
  v=0
  for i in range(t):
    plt.scatter(w0,a*w0*w0*w0*w0 + b*w0*w0*w0 + c*w0*w0 + d*w0 + e,color='black')
    dL = 4*a*w0*w0*w0 + 3*b*w0*w0 + 2*c*w0 + d
    v=(1-beta)*v + beta*dL
    w0=w0 - lr*v
  plt.scatter(w0,a*w0*w0*w0*w0 + b*w0*w0*w0 + c*w0*w0 + d*w0 + e,color='yellow')
```

```python
grades(1/5,-1/2,-1,1/4,0,0.1,1000)
```

```python
moment(1/5,-1/2,-1,1/4,0,0.1,0.05,1000)
```

Normal gradient descent converged to a minima but it was not a global minima whereas gradient descent with momentum converged to global minima.

```python
def gra_des(a,b,c,d,e,lr1,lr2,t):
  x = np.linspace(-20, 20, 10) 
  y = np.linspace(-20, 20, 10) 
  [X, Y] = np.meshgrid(x, y) 
  fig, ax = plt.subplots(1, 1) 
  z = np.array(a*X*X + b*Y*Y + c*X + d*Y + e)
  ax.contourf(X, Y, z, 100, cmap = 'jet') 
  ax.axis('scaled')
  ax.set_title('Contour Plot') 
  ax.set_xlabel('x') 
  ax.set_ylabel('y')
  w1=15
  w2=15
  for i in range(t):
    plt.scatter(w1,w2,color='black')
    dx=2*a*w1 + c
    dy=2*b*w2 + d
    w1=w1-lr1*(dx)
    w2=w2-lr2*(dy)
  plt.scatter(w1,w2,color='yellow')
```

```python
gra_des(1/2,1/2,0,0,0,1,1000)
```

```python
gra_des(1/2,1/2,0,0,0,0.2,1.8,50)
```
