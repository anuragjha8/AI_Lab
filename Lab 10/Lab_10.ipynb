```python
import numpy as np
import math
```

```python
def pout(w,a):
  pmin=[0.01,0.02,0.03,0.1,0.3]
  pmax=[0.1,0.2,0.3,0.5,0.7]
  out=pmin[a]+((pmax[a] - pmin[a])*(w-1)/9)
  return out
```

```python
bestScore = np.zeros((301,11))
shot = np.zeros((301,11))
act = [1,2,3,4,6]

for b in range(1,301):
  for w in range(1,11):
    for i in range(5):
      hit=act[i]
      out = pout(w,i)
      if(b+10-w<=300):
        runs=(1-out)*(hit + bestScore[b-1][w]) + out*(bestScore[b-1][w-1])
        if(runs > bestScore[b][w]):
          bestScore[b][w] = runs
          shot[b][w] = hit
```

```python
for i in range(1,301):
  for j in range(1,11):
    if(i+10-j<=300):
      print('runs possible when ',i,' balls remaining and ',j,' wickets remaining',bestScore[i][j])
```
