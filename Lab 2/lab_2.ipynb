```
import numpy as np
import matplotlib.pyplot as plt
xs=int(input("Enter rows "))
ys=int(input("Enter columns "))

rows, cols = (10, 3) 
arr = np.zeros((rows,cols))
b = np.zeros((100,2))

for i in range(10):
  x=np.random.randint(xs)
  y=np.random.randint(ys)
  arr[i][0]=x
  arr[i][1]=y
  arr[i][2]=10
print(arr)
```

```
xt=int(input("Current x "))
yt=int(input("Current y "))
score=0
```

```

for i in range(100):
  r=np.random.randint(5)
  if r==0:
    m="up"
    yt=yt+1
    if yt>=ys:
      score=score-10
      yt=ys-1

  if r==1:
    m="down"
    yt=yt-1
    if yt<0:
      score=score-10
      yt=0

  if r==2:
    m="left"
    xt=xt-1
    if xt<0:
      score=score-10
      xt=0

  if r==3:
    m="right"
    xt=xt+1
    if xt>=xs:
      score=score-10
      xt=xs-1

  if r==4:
    c=0
    m="pick"
    for j in range(10):
      if (arr[j][0]==xt and arr[j][1]==yt):
        c=1
        score=score+arr[j][2]
        arr[j][2]=0
    if c==0:
      score=score-1

  b[i][0]=score
  b[i][1]=(score/(i+1))

  print("time: ",i+1," movement ",m," current cooredinate (",xt,",",yt,") "," current score: ",score)
  for i in range(10):
    if arr[i][2]==0:
      print("NO",end=" ")
    else:
      print("YES",end=" ")
  print()

```

```
for i in range(100):
  print("current average is ",b[i][1])
```

```
plt.plot(b[:, 1], 'r', label = 'performance')
plt.title('average score plot')
plt.xlabel('Time')
plt.ylabel('Average score')
plt.legend()
```
