Question 1

```
import numpy as np
```

```
fil = open('speechs.txt','r')
txt=fil.read()
txt=txt.lower()
wor=txt.split()
```

```
speech=[]
for i in wor:
  if i not in speech:
    speech.append(i)
print(speech)
```

```
for i in speech:
  print(i,": ",wor.count(i))
```

Question 2

```
l=len(speech)
print("length of speech ",l)
dic={}
c=0
for i in speech:
  dic[i]=c
  c=c+1
f=np.zeros((l,l))
for i in range(0,len(wor)-1):
  x=dic[wor[i]]
  y=dic[wor[i+1]]
  f[x][y]=f[x][y]+1   
#for i in range(0,l):
#  for j in range(0,l):
#    print(f[i][j],end=" ")
#  print()                                    
```

```
inp=input("Enter word ")
x=dic[inp]
for i in range(0,l):
  if f[x][i]>0:
    print(speech[i],": ",f[x][i])
```
