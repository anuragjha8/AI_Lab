```python
import numpy as np
from numpy import genfromtxt
from numpy import linalg
import matplotlib.pyplot as plt
dat = genfromtxt('Dataset.csv', delimiter = ',', dtype = float)
print(dat)
```

```python
k=int(input("Enter K: "))
centroid=np.zeros((k,2))
for i in range(k):
  centroid[i][0]=np.random.uniform(low=-5,high=5)
  centroid[i][1]=np.random.uniform(low=-5,high=5)
print("Initial centroids are: ")
print(centroid)
ssse=np.zeros(10)
```

```python
def my_function(dat,k,centroid,clus):
  n=len(dat)
  for r in range(10):
    for j in range(n):
      m=10000
      x=0
      for q in range(k):
        d=np.linalg.norm(centroid[q, :] - dat[j, :])
        if d<m:
          x=q
          m=d
      clus[j]=x

    for i in range(k):
      c=0
      sx=0
      sy=0
      for j in range(n):
        if clus[j]==i:
          c=c+1
          sx=sx+dat[j][0]
          sy=sy+dat[j][1]
      if c!=0:
        centroid[i][0]=sx/c
        centroid[i][1]=sy/c
  
    sse=0
    for i in range(n):
      sse+=(np.linalg.norm(centroid[int(clus[i]),:]-dat[i,:])**2)
    #print("SSE this iteration is: ",sse)
    ssse[r]=sse
  '''print("The cluster number associated with each point is:")
  print(clus)
  print("The new centroids are: ")
  print(centroid)
  print("The SSE is: ",sse)'''
```

```python
def part(dataset,clus,k):
  lst=[]
  for i in range(k):
    li=[]
    for j in range(len(clus)):
      if(clus[j]==i):
        li.append([dataset[j][0],dataset[j][1]])
    li=np.array(li)
    lst.append(li)
  lst=np.array(lst)
  return lst
```

```python
n=len(dat)
clus=np.zeros(n)
my_function(dat,k,centroid,clus)
```

```python
k=20
datz=np.random.uniform(low=-1,high=1,size=(10000,2))
centr=np.zeros((k,2))
for i in range(k):
  centr[i][0]=np.random.uniform(low=-1,high=1)
  centr[i][1]=np.random.uniform(low=-1,high=1)
clus=np.zeros(len(datz))
my_function(datz,k,centr,clus)
c_dist=part(datz,clus,k)
for i in range(k):
      if len(c_dist[i])!=0:
        plt.scatter(c_dist[i][:,0], c_dist[i][:,1])
plt.scatter(centr[:, 0], centr[:, 1],marker='+',color='black')
```
