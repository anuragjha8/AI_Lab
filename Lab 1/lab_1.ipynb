```
n = int(input('Enter number of points '))
p=np.random.uniform(low = -1, high = 1, size = (n, 2))
#print(p)
a=[]
for i in range(n):
  s=(p[i][0],p[i][1])
  a.append(s)
A=np.array(a)
print(A)
plt.scatter(p[:, 0], p[:, 1],color='red')
```

```
x1 = int(input("Enter new x "))
y1 = int(input("Enter new y "))
def clos(p):
  p1=np.array((x1,y1))
  p2=np.array((p[0],p[1]))
  d=np.linalg.norm(p1-p2)
  return d

a.sort(key=clos)
k=int(input('enter k: '))
arr=[]
for i in range(k):
  arr.append((a[i][0],a[i][1]))

print(k,' closest points to new point are: ')
print(arr)
ar=np.array(arr)
plt.scatter(ar[:, 0], ar[:, 1],color='red')
plt.scatter(x1,y1)
```

```
def ang(poi):
  x1 = int(input("Enter new x "))
  y1 = int(input("Enter new y "))

  p = np.array((x1, y1))
  arr = []
  for i in poi:
    if np.dot(p, i)>0:
      arr.append(i)

  arr = np.array(arr)
  return arr

z=ang(p)
print('points that make positive angle with new points are:')
print(z)
```

```
b=np.random.uniform(-0.5,0.5,size=100)
k=int(input('Enter k'))
s=0
mean=[]
for i in range(1,k+1):
  s=s+b[i]
  mean.append((s/i))
```

```
w=np.linspace(1,100,num=100)
L=[]
L1=[]
for i in range(len(w)):
  L.append(w[i]**(-1/2))
  L1.append(-1*w[i]**(-1/2))
plt.plot(mean)
plt.plot(w,L,color='red')
plt.plot(w,L1,color='black')
```
